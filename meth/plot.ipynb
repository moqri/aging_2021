```python
import pandas as pd
import pyBigWig
import numpy as np
import seaborn as sns
```

```python
ezs=pd.read_csv('ezs_blood',index_col=0)
ezs.index=range(1,23)
cols=['iPS','ESC','New Born','Centenerian']
#cols=['iPS','ESC','Young','Old']
ezs.columns=cols+['','']
ezs=ezs.div(ezs.ESC,0)
myorder = [1,2,3,0]
ezs['chr']=ezs.index
d=ezs[[cols[i] for i in myorder]+['chr']].melt('chr')
ax=sns.barplot(data=d,x='chr',y='value',hue='variable')
ax.set_xlabel("Chromosome")
ax.set_ylabel("Metylation Level");
ax.set_ylabel("Chromosome-level aging biomarker");
ax.set_title('Whole-Genome Epigenetic Marker, Blood, Newborn vs Centenarian');
```
