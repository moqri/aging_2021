```python
import pandas as pd
import pyBigWig
import numpy as np
import seaborn as sns
sns.set(rc={'figure.figsize':(20,5)})
import matplotlib.pylab as plt
```

```python
ezs=pd.read_csv('ezs_blood',index_col=0)
ezs.index=range(1,23)
cols=['iPS','ESC','New Born','Centenerian']
#cols=['iPS','ESC','Young','Old']
ezs.columns=cols+['','']
ezs=ezs.div(ezs.ESC,0)
myorder = [1,2,3,0]
ezs['chr']=ezs.index
d=ezs[[cols[i] for i in myorder]+['chr']].melt('chr')
d.columns=['chr','Cell Type','value']
ax=sns.barplot(data=d,x='chr',y='value',hue='Cell Type')
ax.set_xlabel("Chromosome",fontsize=16);
ax.set_ylabel("Metylation Level");
ax.set_ylabel("Chromosomeal aging index",fontsize=16);
#ax.set_title('Epigenetic aging index - newborn vs centenarian CD4+ T cells');
plt.setp(ax.get_legend().get_texts(), fontsize='16') # for legend text
plt.setp(ax.get_legend().get_title(), fontsize='20') # for legend title
```
