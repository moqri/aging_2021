```python
import pandas as pd
import seaborn as sn
bl_file='https://ftp.ncbi.nlm.nih.gov/geo/series/GSE40nnn/GSE40279/matrix/GSE40279_series_matrix.txt.gz'
li_file='https://ftp.ncbi.nlm.nih.gov/geo/series/GSE48nnn/GSE48325/matrix/GSE48325_series_matrix.txt.gz'
sc_file='https://ftp.ncbi.nlm.nih.gov/geo/series/GSE31nnn/GSE31848/matrix/GSE31848_series_matrix.txt.gz'
```

```python
def get_data():
    !wget -q -nc {sc_file} -O ../data/sc.txt.gz
    !gzip -d ../data/sc.txt.gz
    !wget -q -nc {bl_file} -O ../data/bl.txt.gz
    !gzip -d ../data/bl.txt.gz    
    !wget -q -nc {li_file} -O ../data/li.txt.gz
    !gzip -d ../data/li.txt.gz
#get_sc()
```

```python
bl=pd.read_table('../data/bl.txt',skiprows=69,index_col=0)
bl_ages=pd.read_table('../data/bl.txt',skiprows=33,nrows=0,index_col=0).columns.values
bl_ages=[float(age.strip('age: ').split('y')[0]) for age in bl_ages]
bl=bl.T.drop('!series_matrix_table_end',1)
bl['age']=bl_ages
bl.age.hist(bins=50)
```

```python
cor=bl.corrwith(bl.age)
p50=cor.drop('age').sort_values().tail(50)
n50=cor.sort_values().head(50)
```

```python
sc=pd.read_table('../data/sc.txt',skiprows=92,index_col=0)
sc_meta=pd.read_table('../data/sc.txt',skiprows=52,nrows=20,index_col=0)
sc=sc[sc_meta[[c for c in sc_meta.columns if 'ES__' in c]].iloc[0].values].drop('!series_matrix_table_end')
sc.shape
```

```python
bll=bl[bl.age<bl.age.quantile(1/3)].median()
blh=bl[bl.age>bl.age.quantile(2/3)].median()
blm=bl.median()
bl=pd.concat([bll, blm, blh], axis=1)
bl.columns=['bll','blm', 'blh']
```

```python
df=bl.drop('age')
df['sc']=sc.median(1).loc[df.index]
```

```python
data=df.loc[n50.index][['sc','bll','blh']]
data.columns=['Stem cells','Young cells','Old cells']
sn.lineplot(data=data.T,legend=False)
```

```python
data['up']=data['sc']<data['blm']
data['down']=data['sc']>data['blm']
data.sum()[['up','down']].plot(kind='bar')
```

```python
data=df.loc[p50.index]
sn.lineplot(data=data[['sc','bll','blh']].T,legend=False)
```

```python
data['up']=data['sc']<data['blm']
data['down']=data['sc']>data['blm']
data.sum()[['up','down']].plot(kind='bar')
```

```python
li=pd.read_csv('../data/li.txt', index_col=0,sep='\t', skiprows=60)
li=li.drop('!series_matrix_table_end')
li_ages=pd.read_csv('../data/li.txt', index_col=0,sep='\t', skiprows=27,nrows=9).iloc[8].str.strip('age: ').astype(float)
li=li.T
li['age']=li_ages
li.age.hist(bins=50)
```

```python
cor=li.corrwith(li.age)
```

```python
p50=cor.drop('age').sort_values().tail(50)
n50=cor.sort_values().head(50)
```

```python
lil=li[li.age<li.age.quantile(1/3)].median()
lih=li[li.age>li.age.quantile(2/3)].median()
lim=li.median()
li=pd.concat([lil, lim, lih], axis=1)
li.columns=['lil','lim', 'lih']
```

```python
df=li.drop('age')
df['sc']=sc.median(1).loc[df.index]
```

```python
data=df.loc[n50.index]
```

```python
sn.lineplot(data=data[['sc','lil','lih']].T,legend=False)
```

```python
data=df.loc[n50.index]
data['up']=data['sc']<data['lim']
data['down']=data['sc']>data['lim']
data.sum()[['up','down']].plot(kind='bar')
```

```python
data=df.loc[p50.index]
sn.lineplot(data=data[['sc','lil','lih']].T,legend=False)
```

```python
data['up']=data['sc']<data['lim']
data['down']=data['sc']>data['lim']
data.sum()[['up','down']].plot(kind='bar')
```
